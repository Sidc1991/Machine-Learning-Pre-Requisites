# How Do I Filter Rows of Pandas DataFrame by Column Values

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.shape
```

```python
#Suppose we want to filter our dataframe and see rows where duration is >= 200 minutes.
#We are going to start off with the old fashioned way of building a for loop

booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
#Creating a list of series of long movies called is_long

is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
#We will now pass is_long into the movies data frame to pick out movies we want
#Notice the type of brackets used
movies[is_long]
```

### This was the long way to get to the results. Now for the shortcut.

In most cases we don't need to write the for loop to get the results.

```python
#Pandas understands that you want to look at the series(columns) duration and return
#true false statements that show rows that have values greater or less than specified amount.
is_long = movies.duration >= 200
```

```python
is_long.head()
```

#### Pass the above is_long into movies to return the filtered dataframe. Main take away: filter using booleans.

```python
#movies[movies.duration >= 200] works as well
movies[is_long]
```

```python
#if i want to return specific series along with my filtered columns I use the .loc command
movies.loc[movies.duration >= 200, 'genre']
```

# How Do I Apply Multiple Filter Criteria to a Pandas DataFrame **

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

In the last section we only looked at single filter criterion, i.e. movies with duration >= 200. In this section I want to expand the scope to include multiple filter criteria. We achieve this using the 'and' and 'or' operators.

Conceptually we want: movies[movies.duration >= 200 and movies.genre == 'Drama']

However, we need to make some tweaks to the above.

First:
**Adding parenthesis** around our criteria will preserve the order.

Second:
**Can't use 'and'** operator. Instead, we have to use **&** operator.

Third:

**Can;t use 'or'** operator. Instead, we have to use **|** operator.

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
tester = movies[(movies.genre == 'Drama')|(movies.genre == 'Crime')]
```

```python
#If i wanted to sort out my new data frame that I have created above I
#assign it to an object tester and then run sort values on a particular series.

#tester.title.sort_values()
```

#### A more simpler way to generate an or function is using the 'isin' command.

```python
movies.genre.isin(['Crime', 'Drama', 'Action'])

#The above is command is identical to movies[(movies.genre == 'Drama')|(movies.genre == 'Crime')]
```

# Why Should I Use a 'Groupby' in Pandas?

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
#If I wanted mean of drinks by beer servings

drinks.beer_servings.mean()
```

```python
#Now suppose I wanted to group the mean by continents:**
drinks.groupby('continent').beer_servings.mean()
```

```python
#Picking specific continent
drinks[drinks.continent == 'Africa'].beer_servings.mean()
```

```python
#Mean is not the only aggregation function we can use. We can use max and min as well:
drinks.groupby('continent').beer_servings.max()
```

#### A more useful function than max, mean is the agg function which lets us aggregate the different results into a new dataframe **

```python
drinks.groupby('continent').beer_servings.agg(['count', 'min', 'max', 'mean'])
```

```python
#If we want mean of all columns

drinks.groupby('continent').mean()
```

```python
#It is easy to visualize what's going on above with matplotlib.
%matplotlib inline
```

```python
drinks.groupby('continent').mean().plot(kind = 'bar')
```

# How Do I handle Missing Values in Pandas?

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
#Let's take a look at the tail instead of head, i.e. the last 5 rows.

ufo.tail()
```

```python
# We see NaN (not a number) in colors reported column. It conceptually means a missing value.
#We can get a better view of missing values using the isnull() command where True means the value is missing and False
#means the value exists.

ufo.isnull().tail()
```

```python
#The inverse method is notnull() and yields opposite of above table
ufo.notnull().tail()
```

```python
#To get total number of missing values in each of the columns:

ufo.isnull().sum()

#We can get a sum of booleans because pandas converts True to 1 and False to 0.
```

```python
#To see all the missing values in dataframe we can specify the column and get the results

ufo[ufo.City.isnull()]
```

```python
#One common thing to do with missing values is to drop them.
ufo.shape
```

```python
#Drop a row (with na) if 'any' of its values are missing

ufo.dropna(how = 'any').shape

#When we do that only 2486 of 18241 rows remain
```

```python
#We can change how to "all" which drops a row only if ALL the values are missing

ufo.dropna(how = 'all').shape

#No rows were dropped since state and time have 0 missing values
```

```python
#We can now specify to drop rows if any (or all) specific columns are missing values

ufo.dropna(subset = ['City', 'Shape Reported'], how = 'any').shape
```

```python
#FILLING MISSING VALUES:

ufo['Shape Reported'].value_counts(dropna = False)

#What if I want to fill the 2644 NaN values in.
```

```python
#To make the changes I use the following command

ufo['Shape Reported'].fillna(value = 'VARIOUS', inplace = True)

#inplace ensures the NaN are replaced.
```

```python
ufo['Shape Reported'].value_counts(dropna = False)
```

# How Do I Work With Dates and Times in Pandas?

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
#What if I wanted to analyze these sightings by time of day or date

ufo.dtypes

#WE see all columns are stored as strings.
```

```python
#We see the time is stored as an object i.e. string. So if I want to extract the hours from the string I would need
#to do some string slicing.

ufo.Time.str.slice(-5,-3).astype(int).head()

#slicing from -5 characters from the end to -3. We will also convert this into integer using astype(int) command
```

```python
#The above applied method is brittle and could break for certain cases of dataframes. A better way is using pandas
#data and time format. 

ufo['Time'] = pd.to_datetime(ufo.Time)
```

```python
ufo.dtypes

#We see last column has changed to date-time.
```

```python
#The pandas datetime can figure out exactly hour, day, week etc it was when the event occured
#Go to .dt. on pandas website to get more commands
ufo.Time.dt.hour.head()
```

```python
ufo.Time.dt.dayofyear.head()
```

```python
#Creating timestamp for comparisons
ts = pd.to_datetime('1/1/1999')
```

```python
#A time stamp (ts) comparison can be the following

ufo.loc[ufo.Time >= ts, :].head()

#The above code is only showing us sightings from dates equal to or greater than our time stamped day
```

```python
#To get the latest time in time series

ufo.Time.max()
```

```python
#We can even see the number of days between the first and last time entry.
ufo.Time.max() - ufo.Time.min()
```

```python
#If we only want days

(ufo.Time.max() - ufo.Time.min()).days
```

```python
%matplotlib inline
```

```python
#Add a year column
ufo['Year'] = ufo.Time.dt.year
```

```python
#Number of sightings by year plotted
ufo.Year.value_counts().sort_index().plot()
```
